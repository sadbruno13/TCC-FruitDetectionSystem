```python
import base64
from matplotlib import pyplot as plt
import matplotlib.image as mpimg

def image_to_base64(image_path):
    with open(image_path, "rb") as image_file:
        encoded_string = base64.b64encode(image_file.read()).decode('utf-8')
    return encoded_string

def plot_image(image_path, title_name="Imagem Original"):
    # Carrega a imagem usando matplotlib
    img = mpimg.imread(image_path)
    
    # Cria a figura e os eixos
    fig, ax = plt.subplots(figsize=(10, 10))
    
    # Plota a imagem original
    ax.imshow(img)
    ax.axis('off')  # Remove os eixos
    ax.set_title(title_name)
    
    plt.show()

img_file_path = r"E:\Repositorios\TCC-FruitDetectionSystem\api\images\20240503_181201.jpg"

imageb64 = image_to_base64(img_file_path)
plot_image(img_file_path)

headers = {
        'Content-Type': 'application/json',
    }
data = {
    'imageb64': imageb64
}
```

```python
import requests

url = "http://192.168.15.58:5000/score"  # Substitua pela URL da API que voc deseja acessar

response = requests.post(url, json=data, headers=headers)

if response.status_code == 200:
    data = response.json()
    print(data)
    # Faa algo com os dados retornados pela API
else:
    print("Erro na requisio:", response.status_code)

plot_image(img_file_path, title_name=data["Resultado"])
```
